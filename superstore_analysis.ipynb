```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Himangi\OneDrive\Documents\SuperStore Data.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# Total Sales
```

```python
print(df["Sales"].sum())
```

```python
# Total Profit
```

```python
print(df["Profit"].sum())
```

```python
# Sales by Category
```

```python
print(df.groupby ("Category")["Sales"].sum())
```

```python
# Loss-Making Sub-Categories
```

```python
loss = df. groupby ("Sub-Category")["Profit"].sum()
print(loss[loss < 0])
```

```python
# Total Quantity
```

```python
print(df["Quantity"].sum())
```

```python
# Top 5 States
```

```python
print(df.groupby("State")["Sales"].sum().sort_values(ascending=False).head())
```

```python
# Profit Margin Column
```

```python
df["Profit_Margin"] = (df["Profit"] / df["Sales"]) * 100
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
#Sales by Category # Charts
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
# Profit by Region
```

```python
region_profit = df.groupby("Region")["Profit"].sum()

region_profit = region_profit[region_profit > 0]

region_profit.plot(kind="pie", autopct="%1.1f%%")
plt.title("Profit by Region")
plt.show()
```
